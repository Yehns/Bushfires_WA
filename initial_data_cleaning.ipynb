```python
import json
import pandas as pd
```

```python
with open('DBCA_Fire_History_DBCA_060_WA_GDA2020_Public.geojson') as f:
    data = json.load(f)
```

```python
d1 = pd.DataFrame(data["features"])
d1
```

```python
d3 = d1['properties'].apply(pd.Series)
d3.head()
```

```python
d4 = d1['geometry'].apply(pd.Series)
d4.head()
```

```python
d4['coords'] = d4.coordinates.apply(lambda x:x[0][0])
d4.head()
```

```python
df2 = d4.drop('coordinates', axis=1)
df2.head()
```

```python
df = pd.concat([df2, d3], axis=1)
df.head()
```

```python
df.info
```

```python
df['lat'] = df.coords.apply(lambda x:x[0])
df['lon'] = df.coords.apply(lambda x:x[1])
df.head()
```

```python
df = df.drop('type', axis=1)
```

```python
df.to_csv('bushfire_data2.csv', sep=',')
```
